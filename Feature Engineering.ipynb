<b> Creating a Basic Logistic Regression Model </b>

```python
# Splitting the dataset. taking the label from both test and train and treating it to be 'y' the target variable 
y_train= train['labels']
y_test= test['labels']
X_train=train.drop(['labels'],axis=1)
X_test=test.drop(['labels'],axis=1)
```

```python

###-------------------------Logistic Regression before feature elimination-----------------------
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score




#y_train=y_train.values.reshape(-1,1)
classifier = LogisticRegression()
classifier.fit(X_train, y_train.values.ravel())
##Here values.ravel() is nothing but values.reshape(-1,)
y_pred1 = classifier.predict(X_train)
k=accuracy_score(y_train,y_pred1)
print("The Score for this model for train data is  :",k)
y_pred2 = classifier.predict(X_test)
k1=accuracy_score(y_test,y_pred2)
print("The Score for test data is : ",k1)







```

```python
For the above basline model the accuracy of the logistic regression model is 59%. Now we will do feature engineering 
and try to increase the accuracy of the model.
```

For feature elimination I have used is "Select K Best". This algorithm basically scores all the features using some
inbuilt function in my case i have used f_classif to score all the variables after scoring it given out k variables 
having best scores. We need to specify how many k values we want in my case i have selected 22 best features so the 
algorithm gives out 22 best features from the set of all features. The funtion we used for scoring f_classif is nothing 
but ANOVA F-value between feature for classification tasks. 

   


<b> Performance of Logistic Regression After Feature Engineering </b>

```python


#-------------------------------Feature Elimination------------------------------------
from sklearn.feature_selection import SelectKBest, f_classif
names =list(range( 0, 500))
X, y = X_train, y_train
X.columns = names

#Selecting 22 features
p=SelectKBest(f_classif, k=22)
X_new= p.fit_transform(X_train, y_train.values.reshape(-1,))
X_new=pd.DataFrame(X_new)
new_cols = X.columns.values[p.get_support()]


#Updating our X_train and X_test with feature elimination
new_X_train= X_train.iloc[:,new_cols]
new_X_test=X_test.iloc[:,new_cols]




#---------------------------------Logistic regression after feature elimination--------------------------

classifier = LogisticRegression()
classifier.fit(new_X_train, y_train.values.ravel())
y_pred2 = classifier.predict(new_X_test)
k1=accuracy_score(y_test,y_pred2)
print("Accuracy of logistic regreesion after feature elimination: ",k1)
```

So before we did feature elimination the accuracy of logistic regression was 59% for test data. After we did feature 
elimination now we can see that the accuracy of logistic regreesion is 62%. We can clearly see an increase in the 
accuracy of the test data after feature engineering.
